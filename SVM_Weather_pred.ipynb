### **Removing outliers**

```python
weather_standardized
```

```python
import numpy as np

# Defining a function to handle outliers using mean imputation
def impute_outliers(data, threshold=3):
    imputed_data = data.copy()


    for col_idx in range(imputed_data.shape[1]):

        z_scores = np.abs((imputed_data[:, col_idx] - np.mean(imputed_data[:, col_idx])) / np.std(imputed_data[:, col_idx]))   # Calculating  the z-scores for each data point


        outliers = z_scores > threshold # Identify outliers based on the threshold
        imputed_data[outliers, col_idx] = np.mean(imputed_data[:, col_idx])


    return imputed_data

imputed_array = impute_outliers(weather_standardized,threshold=3)
```

 # **Training Models**

```python
from sklearn import svm # Importing the Support Vector Machine (SVM) model from scikit-learn
from sklearn.model_selection import train_test_split# Importing the train_test_split function from scikit-learn to split data into training and test sets
from sklearn.preprocessing import StandardScaler # Importing the StandardScaler from scikit-learn to standardize features by removing the mean and scaling to unit variance
```

### **Training the model using SVM**

```python
target=weather['weathercode']
target
```

## **Data Split**

```python
scaler = StandardScaler() # Initializing the scaler

features = scaler.fit_transform(features)  # Standardizing the features

features_train, features_test, target_train, target_test = train_test_split(features, target, test_size=0.3, random_state=42) # Splitting the data into training and test sets
```

### **Managing Imbalanced Data**

```python
from sklearn.metrics import  precision_score, recall_score, f1_score, log_loss, accuracy_score # Importing performance measuring libraries
```

**Managing imbalanced data using SVM parameter and training SVM model**

```python

model = svm.SVC(kernel='linear', class_weight='balanced', random_state=42)  # Defining svm model to balance data
model.fit(features_train, target_train)
predictions_svm_bal = model.predict(features_test) # Calculating the predictions for balaced data


accuracy_svm_bal = accuracy_score(target_test, predictions_svm_bal) # Calculating the accuracy for balaced data

print('SVM Model Accuracy for balanced data by SVM parameter:', accuracy_svm_bal)

```

```python
decision_function_output = model.decision_function(features_test)

probability_predictions_svm_bal = (np.exp(decision_function_output).T / np.exp(decision_function_output).sum(axis=1)).T # Converting decision_function_output to probability estimates

precision_svm_bal = precision_score(target_test, predictions_svm_bal, average='weighted') # Calculating Precision

recall_svm_bal= recall_score(target_test, predictions_svm_bal, average='weighted') # Calculating Recall

f1_svm_bal = f1_score(target_test, predictions_svm_bal, average='weighted') # Calculating F1-Score

logloss_svm_bal = log_loss(target_test, probability_predictions_svm_bal)  # Use probability predictions

print("SVM Precision for balanced data:", precision_svm_bal)
print("SVM Recall for balanced data:", recall_svm_bal)
print("SVM F1-Score for balanced data:", f1_svm_bal)
print("SVM Multiclass Log Loss for balanced data:", logloss_svm_bal)
```

**Managing imbalanced data by undersamplying**

```python
from imblearn.under_sampling import RandomUnderSampler # Importing the RandomUnderSampler class from the imbalanced-learn library

res = RandomUnderSampler(random_state=42) # Creating an instance of the RandomUnderSampler class with a fixed random state for reproducibility
features_res, target_res = res.fit_resample(features_train, target_train) # Using the fit_resample method of the RandomUnderSampler instance to undersample the training data
```

```python
class_counts =target_res.value_counts() # Calculate and store the frequency of each unique weather code in the 'target_res' column

class_counts.plot(kind='bar') # Ploting the 'class_counts' as a bar chart

plt.show() # Displaying the plot
```

```python
model.fit(features_res, target_res) # Training the model using  undersampled data

predictions_under_bal = model.predict(features_test) # Using the trained model to make predictions on the test data

accuracy_under_bal = accuracy_score(target_test, predictions_under_bal) # Calculating the accuracy

print('SVM Model Accuracy for undersampled data:', accuracy_under_bal) # Printing the accuracy
```

### **Comparison between the SVM and the Random Forest**

```python
from sklearn import svm
from sklearn.ensemble import RandomForestClassifier

# Assume features is your feature set and target is your target variable
train_sizes = np.arange(0.1, 1.0, 0.1)
accuracies_rf = []
accuracies_svm = []

for size in train_sizes:
    features_train, features_test, target_train, target_test = train_test_split(features_res, target_res, train_size=size, random_state=42)

    # RandomForestClassifier
    model_rf = RandomForestClassifier(bootstrap= False, max_depth= 30, max_features='sqrt', min_samples_leaf= 2, min_samples_split=2, n_estimators= 200)
    model_rf.fit(features_train, target_train)
    predictions_rf = model_rf.predict(features_test)
    accuracy_rf = accuracy_score(target_test, predictions_rf)
    accuracies_rf.append(accuracy_rf)

    # SVM
    model_svm = svm.SVC(C= 10, gamma= 1, kernel= 'linear')
    model_svm.fit(features_train, target_train)
    predictions_svm = model_svm.predict(features_test)
    accuracy_svm = accuracy_score(target_test, predictions_svm)
    accuracies_svm.append(accuracy_svm)

# Plot the results
plt.plot(train_sizes, accuracies_rf, label='Random Forest')
plt.plot(train_sizes, accuracies_svm, label='SVM')
plt.xlabel('Train Size')
plt.ylabel('Model Accuracy')
plt.title('Model Accuracy vs Train Size')
plt.legend()
plt.show()
```

```python

train_sizes = np.arange(0.1, 1.0, 0.1)
f1_scores_rf = []
f1_scores_svm = []
recall_scores_rf = []
recall_scores_svm = []
precision_scores_rf = []
precision_scores_svm = []

for size in train_sizes:
    features_train, features_test, target_train, target_test = train_test_split(features_res, target_res, train_size=size, random_state=42) # Splitting balanced data

    # RandomForestClassifier
    model_rf = RandomForestClassifier(bootstrap= False, max_depth= 30, max_features='sqrt', min_samples_leaf= 2, min_samples_split=2, n_estimators= 200)
    model_rf.fit(features_train, target_train)
    predictions_rf = model_rf.predict(features_test)

    f1_rf = f1_score(target_test, predictions_rf, average='weighted')
    recall_rf = recall_score(target_test, predictions_rf, average='weighted')
    precision_rf = precision_score(target_test, predictions_rf, average='weighted')

    f1_scores_rf.append(f1_rf)
    recall_scores_rf.append(recall_rf)
    precision_scores_rf.append(precision_rf)

    # SVM
    model_svm = svm.SVC(C= 10, gamma= 1, kernel= 'linear')
    model_svm.fit(features_res, target_res)
    predictions_svm = model_svm.predict(features_test)

    f1_svm = f1_score(target_test, predictions_svm, average='weighted')
    recall_svm = recall_score(target_test, predictions_svm, average='weighted')
    precision_svm = precision_score(target_test, predictions_svm, average='weighted')

    f1_scores_svm.append(f1_svm)
    recall_scores_svm.append(recall_svm)
    precision_scores_svm.append(precision_svm)

# Plot the results
plt.figure(figsize=(12, 6))

plt.subplot(1, 3, 1)
plt.plot(train_sizes, f1_scores_rf, label='Random Forest')
plt.plot(train_sizes, f1_scores_svm, label='SVM')
plt.xlabel('Train Size')
plt.ylabel('F1 Score')
plt.title('F1 Score vs Train Size')
plt.legend()

plt.subplot(1, 3, 2)
plt.plot(train_sizes, recall_scores_rf, label='Random Forest')
plt.plot(train_sizes, recall_scores_svm, label='SVM')
plt.xlabel('Train Size')
plt.ylabel('Recall Score')
plt.title('Recall Score vs Train Size')
plt.legend()

plt.subplot(1, 3, 3)
plt.plot(train_sizes, precision_scores_rf, label='Random Forest')
plt.plot(train_sizes, precision_scores_svm, label='SVM')
plt.xlabel('Train Size')
plt.ylabel('Precision Score')
plt.title('Precision Score vs Train Size')
plt.legend()

plt.tight_layout()
plt.show()
```
